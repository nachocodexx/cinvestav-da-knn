a) Escriba un programa para la estimaciÃ³n de densidad kNN de n puntos en d dimensiones, de acuerdo con la EcuaciÃ³n 15, de la clase AD-05.

$p(x) = \frac{k}{nV}$

```python
def knn_px_est(**kwargs):
    # k is the number of the nearest datapoints.
    k = kwargs.get('k',1)
    # N is a constant that represents the total number of patterns in the training set.
    n = kwargs.get('n',N)
#     V = kwargs.get('V',1)
    return lambda V: k/(n*V)
```

$V = V_0 \cdot r^{d}$

$
V_0 =
\begin{cases}
\pi^{d/2} / (d/2)!, & \text{si }d\text{ es par} \\
2^{d} \pi^{d-1}/2 (\frac{d-1}{2}!) / d!, & \text{si }d\text{ es impar}
\end{cases}
$

```python
def knn_V(r,**kwargs):
    __factorial = np.math.factorial
    # D is a constant that represent the number of dimensions
    d   = kwargs.get('d',D)
    fn1 = np.power(np.pi,d//2)/__factorial(d//2)
    
    fn2 = (np.power(2,d) * np.power(np.pi,(d-1)//2)*__factorial((d-1)//2))/__factorial(d)
    Vo  = fn1 if(d%2 ==0) else fn2
    return Vo*np.power(r,d)
```

```python
n = 100
mu, sigma = 10, 3 # mean and standard deviation
s = np.random.normal(mu, sigma, n)
```

```python
def zipM(XY):
    for xs,ys in XY:
        for x in list(zip(xs,ys)):
            yield x
```

```python
def L(x,u,sigma):
    value =  np.sum(list(map(lambda xk: (xk - u)**2 ,x)))
    return -(n*.5)*np.log(2*np.pi*sigma) - (1/(2*sigma))*value
#     print(value)
# L(s,y,sigma**2)
xLen = np.linspace(1, 20,10)
yLen = np.linspace(1, 20,10)
X,Y = np.meshgrid(xLen,yLen)
XY  = np.array(list(zip(X,Y)))
result=[]
for  uk,sigma in zipM(XY):
    lTheta = L(s,uk,sigma)
    result.append(lTheta)
Z = np.array(result).reshape(X.shape[0],Y.shape[0])
```

```python
def windowFn(xs,unit_len=1):
    return np.invert(np.any( np.abs(xs) > (unit_len/2 ), axis=1 )).astype(int)

    
def fkn(X):
    return np.sum(windowFn(X))
#  P^
def parzenWindowEst(**kwargs):
    k  = kwargs.get('k',1)
    V  = kwargs.get('V',1)
    n  = kwargs.get('n',1)
    return (k/n*V)
    
    
def px_est(X,**kwargs):
    n = X.shape[0]
    h = kwargs.get('h',1)
    V = kwargs.get('V',1/np.sqrt(n))
    x = kwargs.get('center',np.zeros(X.shape[1]))
    k = np.sum(windowFn(distance1(x)(X)/h))
    p = parzenWindowEst(k=k,V=V,n=n)
    return p

center = np.array([0,0,0])
px_est(
    X,
    V      = 1,
    h      = 1,
)
```

```python
def gaussianKernel(X,**kwargs):
    center = kwargs.get('center',np.zeros(X.shape[1]))
    h      = kwargs.get('h',1)
    d      = X.shape[1]
    u      = distance3(center)(X)/h 
    exp0   = 1/(( np.sqrt(2*np.pi))**d)
    exp1   = np.exp(-.5*(u**2))
    return exp0*exp1
def px_gaussian(X,**kwargs):
    V      = kwargs.get('V',1)
    center = kwargs.get('center',np.zeros(X.shape[1]))
    n      = X.shape[0]
    h      = kwargs.get('h',1)
    _X      = gaussianKernel(X,h=h,center=center)
    return np.sum((1/V)*(_X))/n
```
