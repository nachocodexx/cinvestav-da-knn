```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
df = pd.read_csv('data.csv',delimiter='\t',header=None)
X  = df.values
df
```

```python
w0 = X[:,:3]
w1 = X[:,3:6]
w2 = X[:,6:]
```

```python
%matplotlib notebook
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
ax.scatter(w0[:,0],w0[:,1],w0[:,2],color="black",s=100,alpha=1)
ax.scatter(w1[:,0],w1[:,1],w1[:,2],color="red",s=100,alpha=1)
ax.scatter(w2[:,0],w2[:,1],w2[:,2],color="blue",s=100,alpha=1)
```

```python
n = 100
mu, sigma = 10, 3 # mean and standard deviation
s = np.random.normal(mu, sigma, n)
```

```python
def L(x,u,sigma):
    value =  np.sum(list(map(lambda xk: (xk - u)**2 ,x)))
    return -(n*.5)*np.log(2*np.pi*sigma) - (1/(2*sigma))*value
#     print(value)
# L(s,y,sigma**2)
xLen = np.linspace(1, 20,10)
yLen = np.linspace(1, 20,10)
X,Y = np.meshgrid(xLen,yLen)

result = []
for i in range(X.shape[0]):
    for j in range(Y.shape[0]):
        uk    = X[i][j]
        sigma = Y[i][j]
        lTheta = L(s,uk,sigma)
        result.append(lTheta)
Z = np.array(result).reshape(X.shape[0],Y.shape[0])
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X,Y,Z,cmap =cm.Dark2_r,linewidth=0,antialiased=True)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
fig.colorbar(surf, shrink=0.5, aspect=20)
```
