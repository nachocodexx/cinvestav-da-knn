```python
n = 100
mu, sigma = 10, 3 # mean and standard deviation
s = np.random.normal(mu, sigma, n)
```

```python
def zipM(XY):
    for xs,ys in XY:
        for x in list(zip(xs,ys)):
            yield x
```

```python
def L(x,u,sigma):
    value =  np.sum(list(map(lambda xk: (xk - u)**2 ,x)))
    return -(n*.5)*np.log(2*np.pi*sigma) - (1/(2*sigma))*value
#     print(value)
# L(s,y,sigma**2)
xLen = np.linspace(1, 20,10)
yLen = np.linspace(1, 20,10)
X,Y = np.meshgrid(xLen,yLen)
XY  = np.array(list(zip(X,Y)))
result=[]
for  uk,sigma in zipM(XY):
    lTheta = L(s,uk,sigma)
    result.append(lTheta)
Z = np.array(result).reshape(X.shape[0],Y.shape[0])
```

```python
fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(X,Y,Z,cmap =cm.Dark2_r,linewidth=0,antialiased=True)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
fig.colorbar(surf, shrink=0.5, aspect=20)
```

```python
def windowFn(xs,unit_len=1):
    return np.invert(np.any( np.abs(xs) > (unit_len/2 ), axis=1 )).astype(int)

    
def fkn(X):
    return np.sum(windowFn(X))
#  P^
def parzenWindowEst(**kwargs):
    k  = kwargs.get('k',1)
    V  = kwargs.get('V',1)
    n  = kwargs.get('n',1)
    return (k/n*V)
    
    
def px_est(X,**kwargs):
    n = X.shape[0]
    h = kwargs.get('h',1)
    V = kwargs.get('V',1/np.sqrt(n))
    x = kwargs.get('center',np.zeros(X.shape[1]))
    k = np.sum(windowFn(distance1(x)(X)/h))
    p = parzenWindowEst(k=k,V=V,n=n)
    return p

center = np.array([0,0,0])
px_est(
    X,
    V      = 1,
    h      = 1,
)
```

```python
def gaussianKernel(X,**kwargs):
    center = kwargs.get('center',np.zeros(X.shape[1]))
    h      = kwargs.get('h',1)
    d      = X.shape[1]
    u      = distance3(center)(X)/h 
    exp0   = 1/(( np.sqrt(2*np.pi))**d)
    exp1   = np.exp(-.5*(u**2))
    return exp0*exp1
def px_gaussian(X,**kwargs):
    V  = kwargs.get('V',1)
    n  = X.shape[0]
    h  = kwargs.get('h',1)
    _X = gaussianKernel(X,h=h)
    return np.sum((1/V)*(_X))/n

data = px_gaussian(X,h=1,V=1)
data
# X.shape
```
